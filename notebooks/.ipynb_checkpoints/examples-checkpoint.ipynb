```python
# import libraries
import numpy as np
from lib import schema as db      # database schema
from datetime import datetime   
from collections import Counter   # count elements in list
from bson.son import SON          # to_bson
from bson.code import Code        # write code for map-reduce
import operator                   # for sorting dict
import matplotlib
import matplotlib.pyplot as plt   # plot plate
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"/usr/share/fonts/truetype/LiHeiPro.ttf", size=11)   # chinese font for matplotlib

# [Convension Functions]
# draw barchart of stat as dict
def dict2barh(dic):
    x = np.arange(len(dic))
    plt.barh(x, dic.values())
    plt.yticks( x + 0.5,  dic.keys() , fontproperties=font)
    plt.show()

# draw barchart of stat as list (sorted dict)    
def list2barh(lis):
    x = np.arange(len(lis))
    plt.barh(x, map(lambda l: l[1], lis))
    plt.yticks( x + 0.5, map(lambda l: l[0], lis), fontproperties=font)
    plt.show()
```

```python
# some simple orm examples (see https://mongoengine-odm.readthedocs.org/guide/querying.html)
print db.campaigns.objects(client_industry='professionalservices').count()   # condition and count
print "-----------------------"
print db.campaigns.objects.first().name     # single attributes
print "-----------------------"
print db.campaigns.objects.first()._data    # all attributes as dict
print "-----------------------"
# collect names
camp = db.campaigns.objects.first()
adgroups = db.adgroups.objects(campaign_id=camp.campaign_id)
print ','.join(map(lambda a: a.name, adgroups))
print "-----------------------"
adgroup_dailys = db.adgroup_dailys.objects(campaign_id=camp.campaign_id)
print adgroup_dailys.count()
spents = map(lambda a: a.spent, adgroup_dailys)
print sum(spents)
```

```python
# collections of custumer verticals (https://mongoengine-odm.readthedocs.org/apireference.html#mongoengine.queryset.QuerySet.item_frequencies)
industries = db.campaigns.objects.item_frequencies('client_industry', normalize=False)   # summarize count as dict
sorted_industries = sorted(industries.items(), key=operator.itemgetter(1))   # sort dict to be sorted-list
list2barh(sorted_industries)  # plot 
```

```python
# aggregation example (by pymongo, http://api.mongodb.org/python/current/examples/aggregation.html)
result = db.adgroup_dailys._get_collection().aggregate([
    { "$group": {"_id": '$client_industry', "ltv": { "$sum" : '$spent' }} },
    {"$sort": SON([("ltv", -1)])}
])['result']

industry_ltv = []
for r in result[:20]: industry_ltv.append([r['_id'], r['ltv']])
list2barh(industry_ltv[0:20])
```

```python
# another aggregation example
result = db.adgroup_dailys._get_collection().aggregate([
    { "$group": {"_id": '$client_name', "ltv": { "$max" : '$lifetime_total_unique_actions' }} },
    {"$sort": SON([("ltv", -1)])}
])['result']

client_ltv = []
for r in result[:20]: client_ltv.append([r['_id'], r['ltv']])
list2barh(client_ltv[:20])
```
